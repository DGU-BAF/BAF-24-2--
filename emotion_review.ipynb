```python
morphes_results = []

# DataFrame의 각 리뷰에 대해 형태소 분석 수행
for review in df['ReviewText']:
    if isinstance(review, str) and review.strip():  # review가 문자열이고 비어있지 않은 경우
        morphes = okt.morphs(review)  # 형태소 분석
        morphes_results.append(morphes)  # 결과를 리스트에 추가
    else:
       morphes_results.append([])  # None 또는 빈 문자열일 경우 빈 리스트 추가
```

```python
def merge_gaseongbi(row):
    # row가 비어 있는 경우 빈 리스트 반환
    if not row:
        return []
    
    merged_row = []
    skip_next = False

    for i in range(len(row) - 1):
        # "가성" 뒤에 "비"가 오는 경우 두 단어를 합침
        if row[i] == "가성" and row[i + 1] == "비":
            merged_row.append("가성비")
            skip_next = True  # 다음 단어는 건너뜀
        elif skip_next:
            skip_next = False
        else:
            merged_row.append(row[i])

    # 마지막 단어 추가
    merged_row.append(row[-1])
    return merged_row
```

```python
def merge_jaegumae(row):
    # row가 비어 있는 경우 빈 리스트 반환
    if not row:
        return []
    
    merged_row = []
    skip_next = False

    for i in range(len(row) - 1):
        # "재" 뒤에 "구매"가 오는 경우 두 단어를 합침
        if row[i] == "재" and row[i + 1] == "구매":
            merged_row.append("재구매")
            skip_next = True  # 다음 단어는 건너뜀
        elif skip_next:
            skip_next = False
        else:
            merged_row.append(row[i])

    # 마지막 단어 추가
    merged_row.append(row[-1])
    return merged_row
```

```python
df['MorphemeResults_modi_gasungbi'] = df['MorphemeResults'].apply(merge_gaseongbi)
```

```python
df.drop(columns=['MorphemeResults'],inplace=True)
```

```python
df['MorphemeResults_modi_jaegumae'] = df['MorphemeResults_modi_gasungbi'].apply(merge_jaegumae)
```

```python
df.drop(columns=['MorphemeResults_modi_gasungbi'],inplace=True)
```

```python
# 예시 stop words
stop_word = ['요', '다', '는', '은', '이', '가', '아직', '에', '게', '과', '입니다','을','를','니','히','리','기','우','구','추','의','감사']
final_morphs = pd.DataFrame(columns=['ProcessedMorphemes'])  # 열 이름을 'ProcessedMorphemes'로 설정

for i in range(len(df)):
    l = df['MorphemeResults_modi_jaegumae'][i]  # 현재 문장의 형태소 리스트
    b = [word for word in l if word not in stop_word]  # 불용어 제거
    final_morphs.loc[i, 'ProcessedMorphemes'] = b  # 결과 리스트를 새로운 DataFrame에 추가

print(final_morphs)
```

```python
df = pd.concat([df, final_morphs], axis=1)
```
