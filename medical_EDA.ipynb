## 전처리

### 수축기혈압, 이완기혈압 결측 처리(평균치 대체) -> 남녀 구분

```python
print(df['수축기혈압'].isnull().sum())
df['이완기혈압'].isnull().sum()
```

```python
male_bloodpressure1_mean = df[df['성별코드'] == 1]['수축기혈압'].mean()
female_bloodpressure1_mean = df[df['성별코드'] == 2]['수축기혈압'].mean()

male_bloodpressure2_mean = df[df['성별코드'] == 1]['이완기혈압'].mean()
female_bloodpressure2_mean = df[df['성별코드'] == 2]['이완기혈압'].mean()
```

```python
# 혈압 결측치를 성별에 따라 평균으로 대체
df.loc[(df['성별코드'] == 1) & (df['수축기혈압'].isnull()), '수축기혈압'] = male_bloodpressure1_mean
df.loc[(df['성별코드'] == 2) & (df['수축기혈압'].isnull()), '수축기혈압'] = female_bloodpressure1_mean

df.loc[(df['성별코드'] == 1) & (df['이완기혈압'].isnull()), '이완기혈압'] = male_bloodpressure2_mean
df.loc[(df['성별코드'] == 2) & (df['이완기혈압'].isnull()), '이완기혈압'] = female_bloodpressure2_mean
```

```python
print(df['수축기혈압'].isnull().sum())
df['이완기혈압'].isnull().sum()
```

#### ROC 분석 (Receiver Operating Characteristic Analysis)
=> 질병과 관련된 이진 타켓변수가 없으므로 질병 여부를 예측할 수 있는 ROC 분석은 불가함.

신장과 체중은 결측치가 없지만, 허리둘레는 56개의 결측치가 있음. -> 허리둘레 결측 대체 후 진행

```python
# 각 열의 NaN 값 개수 확인
#nan_counts = df.isna().sum()
#print(nan_counts) 
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

 # WHtR ≥ 0.5인 사람을 위험군으로 분류
df['high_risk'] = (df['WHtR'] >= 0.5).astype(int)

# 이진 타겟 변수와 예측 변수 설정
y_true = df['high_risk']  # 예: 0 또는 1로 표시된 위험군 여부
bmi_scores = df['BMI']
whtr_scores = df['WHtR']

# NaN 값이 있는 행 제거
df = df.dropna(subset=['high_risk', 'BMI', 'WHtR'])

# 업데이트된 y_true 및 예측 변수 재설정
y_true = df['high_risk']
bmi_scores = df['BMI']
whtr_scores = df['WHtR']

# BMI ROC 커브
fpr_bmi, tpr_bmi, _ = roc_curve(y_true, bmi_scores)
roc_auc_bmi = roc_auc_score(y_true, bmi_scores)

# WHtR ROC 커브
fpr_whtr, tpr_whtr, _ = roc_curve(y_true, whtr_scores)
roc_auc_whtr = roc_auc_score(y_true, whtr_scores)

# ROC 커브 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(fpr_bmi, tpr_bmi, color='blue', lw=2, label=f'BMI (AUC = {roc_auc_bmi:.2f})')
plt.plot(fpr_whtr, tpr_whtr, color='orange', lw=2, label=f'WHtR (AUC = {roc_auc_whtr:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for BMI vs WHtR')
plt.legend(loc='lower right')
plt.grid()  # 그리드 추가
plt.show()
```

- BMI(파란선) : 곡선 아래 면적이 0.94로, 높은 분류 성능을 보임.
    - 모델이 94%의 정확도로 분류할 수 있음을 의미함.
- WHtR(주황선) : 곡선 아래 면적이 1.00으로, 완벽한 분류 성능을 보임.
    - 거의 수평으로 최상단에 위치함(완벽한 민감도와 특이성)
- 대각선 점선 = 무작위 분류기를 의미함. 아래 면적이 0.5인 기준선 역할

- 따라서, WHtR이 BMI보다 더 좋은 예측을 한다는 결론을 도출할 수 있음.

- 하지만, AUC 1.00이 과도하게 높은 값이므로 WHtR의 성능이 현실적인지 검토해봐야 함..

#### 클래스 분류 및 카테고리화


```python
# BMI 카테고리화
def categorize_bmi(bmi):
    if bmi < 18.5:
        return '저체중'
    elif 18.5 <= bmi < 25:
        return '정상'
    elif 25 <= bmi < 30:
        return '과체중'
    else:
        return '비만'
    
# WHtR 카테고리화
def categorize_whtr(whtr):
    if whtr < 0.4:
        return '저체중'
    elif 0.4 <= whtr < 0.5:
        return '정상체형'
    else:
        return '위험체형'
```

```python
df['BMI_category'] = df['BMI'].apply(categorize_bmi)
df['WHtR_category'] = df['WHtR'].apply(categorize_whtr)

df[['BMI_category', 'WHtR_category']]
```
