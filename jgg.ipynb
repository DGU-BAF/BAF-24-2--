```python
supplement_links = []
```

```python
# 페이지 순회하며 링크 수집
while True:
    # 페이지에서 상품 링크 요소 찾기
    products = driver.find_elements(By.CSS_SELECTOR, "a.absolute-link product-link")

    # 각 상품의 링크를 리스트에 추가
    for product in products:
        link = product.get_attribute("href")
        supplement_links.append(link)
    
    # 다음 페이지 버튼 찾기
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, "a.pagination-next")
        if next_button.is_enabled(): #selenium에 있는 메소드 링크가 사용가능한 지 판단
            next_button.click()  # 다음 페이지로 이동
            time.sleep(3)  # 페이지 로딩 대기
        else:
            break  # 더 이상 페이지가 없으면 종료
    except Exception as e:
        print(f"Error navigating to the next page: {e}")
        break
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
# 영양 성분 크롤링 함수
def get_nutrition_facts(link):
    driver.get(link)  # 링크에 접속
    time.sleep(3)  # 페이지 로딩 시간 대기 (필요시 조정)
    
    try:
        # 영양 성분표를 찾음 (가정: table 태그와 특정 클래스가 사용됨)
        nutrition_table = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "div.supplement-fact-container table")))
        
        # 테이블의 각 행을 크롤링
        rows = nutrition_table.find_elements(By.TAG_NAME, "tr")
        
        # 영양성분표 데이터를 담을 딕셔너리
        nutrition_data = {}

        # 각 행에서 데이터를 추출
        for row in rows[3:]:
            columns = row.find_elements(By.TAG_NAME, "td")
            if len(columns) > 1:  # 두 개 이상의 열이 있을 때만 추출
                nutrient_name = columns[0].text  # 성분 이름
                nutrient_value = columns[1].text  # 성분 값 (양)
                #nutrient_percent = columns[2].text # 성분의 하루 권장량에 대한 비율
                #info_nurient = [nutrient_value,nutrient_percent]
                nutrition_data[nutrient_name] = nutrient_value  # 딕셔너리에 저장
                print(nutrition_data)
        
        return nutrition_data  # 딕셔너리 반환

    except Exception as e:
        print(f"Error extracting nutrition facts from {link}: {e}")
        return None
        
```

```python
# 각 링크에서 영양성분표 크롤링
all_nutrition_data = []

for link in supplement_links:
    nutrition_facts = get_nutrition_facts(link)
    if nutrition_facts:  # 데이터를 성공적으로 가져왔을 때만 추가
        all_nutrition_data.append(nutrition_facts)

# 결과 출력
for data in all_nutrition_data:
    print(data)
```
