신장과 체중이 허리둘레와 양의 상관관계를 띄고 있는 것을 확인할 수 있다.

```python
# 신장과 체중을 기준으로 허리둘레의 평균을 그룹별로 계산
grouped_mean = round(dff.groupby(['신장(5cm단위)', '체중(5kg단위)'])['허리둘레'].mean(),2)

grouped_mean
```

```python
grouped_mean = pd.DataFrame(grouped_mean)
grouped_mean = grouped_mean.reset_index()
grouped_mean
#grouped_mean.drop(columns=['index'], inplace=True)
```

```python
# 각 행을 반복하며 처리
for i in range(len(dff)):
    if pd.isna(dff.iloc[i]['허리둘레']):  # 허리둘레가 결측치인 경우
        # 현재 행의 신장과 체중에 맞는 그룹 찾기
        match = grouped_mean[
            (grouped_mean['신장(5cm단위)'] == dff.iloc[i]['신장(5cm단위)']) &
            (grouped_mean['체중(5kg단위)'] == dff.iloc[i]['체중(5kg단위)'])
        ]
        
        if not match.empty:  # match가 비어있지 않은 경우
            dff.at[i, '허리둘레'] = match['허리둘레']  # 값을 대체

dff
        
    
```
