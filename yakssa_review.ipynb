```python
total_review = {}
for i in range(29,len(review_link)):
    driver.get(review_link[i])
    rev = []
    cla7 = driver.find_element(By.CSS_SELECTOR, 'div.categorize-mobile._categorize_mobile > div.detail_review_wrap_mobile > div._detail_review_wrap_mobile > div > span')
    print(cla7.text)
    num = int(cla7.text)
    if(num == 0):
        total_review[f"number_{i}"] = []
    elif(num % 5 == 0):
        num = num//5 + 1
    else:
        num = num//5 + 2
    print(num)

    for page in range(1,num):
        driver.execute_script("SITE_SHOP_DETAIL.changeContentTab('review',{}, 1, 1, 'N', 0);".format(page))
        a= driver.find_element(By.CSS_SELECTOR, 'div.detail_review_wrap_mobile')
        b= a.find_element(By.CSS_SELECTOR, 'div._detail_review_wrap_mobile')
        c= b.find_element(By.CSS_SELECTOR, 'div._review_wrap')
        #print(c.get_attribute('class'))
        d= c.find_element(By.CSS_SELECTOR, 'ul.list_review_wrap')
        e =d.find_elements(By.CSS_SELECTOR, 'li')
        #list_length = len(e)
        for idx in e:
            f= idx.find_element(By.CSS_SELECTOR,'div')
            g= f.find_element(By.CSS_SELECTOR, 'div.text-14')
            h= g.find_elements(By.CSS_SELECTOR, 'span')
            j= h[0].text
            #print(j)
            supplement = f"number_{i}"  # 키 설정
            rev.append(j)
            total_review[supplement] = rev  # 딕셔너리에 키-값 쌍 저장
    print(total_review[supplement])
```
